```python
H = 100
T = 1

Z=[]
#r = [i / 1000 for i in range(5, 55, 5)]
r = list(np.arange(0.005,0.055,0.005))
for i in range(len(r)):
    Z.append(H/((1+r[i])**T))
print(Z)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
ax.plot(r, Z, marker='.', markersize=15)

ax.set_xticks(r)
plt.ylim(93,100)

#タイトル、軸ラベルの設定
ax.set_title('割引債価格', fontsize=18)
ax.set_xlabel('利回り', fontsize=12)
ax.set_ylabel('価格', fontsize=12)
    
ax.legend(fontsize=12)
plt.show()
```

```python
H = 100
r = 0.02

Z=[]
#r = [i / 1000 for i in range(5, 55, 5)]
T = list(np.arange(0.5,5.5,0.5))
for i in range(len(T)):
    Z.append(H/((1+r)**T[i]))
print(Z)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)

ax.plot(T, Z, marker='.', markersize=15)
ax.set_xticks(T)
plt.ylim(86,100)

#タイトル、軸ラベルの設定
ax.set_title('割引債価格', fontsize=18)
ax.set_xlabel('満期までの期間', fontsize=12)
ax.set_ylabel('価格', fontsize=12)
    
ax.legend(fontsize=12)
plt.show()
```
